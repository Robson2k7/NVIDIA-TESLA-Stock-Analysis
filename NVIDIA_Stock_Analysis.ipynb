# AKCJE NVIDIA

```python
import yfinance as yf
import datetime
```

```python
ticker = "NVDA"
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=10*365)
```

```python
nvidia_data = yf.download(ticker, start=start_date, end=end_date)
```

```python
start_price = nvidia_data['Close'].iloc[0]
end_price = nvidia_data['Close'].iloc[-1]
```

```python
percentage_growth = ((end_price - start_price) / start_price)*100
print(f"Procentowy wzrost akcji Nvidia w ostatnich 10 latach wynosi {percentage_growth}%")
```

```python
import yfinance as yf
import datetime
import matplotlib.pyplot as plt

# Set the ticker symbols
ticker_nvda = "NVDA"
ticker_tsla = "TSLA"

# Define the time period (last 10 years)
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=10*365)

# Download the stock data
nvidia_data = yf.download(ticker_nvda, start=start_date, end=end_date)
tsla_data = yf.download(ticker_tsla, start=start_date, end=end_date)

# Plotting the data
plt.figure(figsize=(14, 7))

# NVIDIA stock prices
plt.plot(nvidia_data['Close'], label='NVIDIA', color='blue')

# Tesla stock prices
plt.plot(tsla_data['Close'], label='Tesla', color='red')

# Chart details
plt.title('NVIDIA vs Tesla Stock Prices Over the Last 10 Years')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
